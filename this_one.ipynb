```python
import matplotlib.pyplot as plt
import seaborn as sns


top_20 = spotify_df.sort_values('streams', ascending=False).head(20)

plt.figure(figsize=(12,6))
sns.barplot(x='streams', y='track_name', data=top_20, palette='viridis')

plt.title('Top 20 Spotify Songs by Streams (2023)')
plt.xlabel('Streams')
plt.ylabel('Track Name')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


top_20 = spotify_df.sort_values('streams', ascending=False).head(20)

plt.figure(figsize=(10,6))
sns.scatterplot(data=top_20, x='bpm', y='streams', hue='track_name', palette='tab20', s=100)

plt.title('Streams vs BPM for Top 20 Spotify Songs (2023)')
plt.xlabel('BPM (Beats Per Minute)')
plt.ylabel('Streams')
plt.legend(title='Track Name', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()

```
