```python
import seaborn as sns
import matplotlib.pyplot as plt

# Columns we're analyzing
features = ['in_spotify_playlists', 'in_apple_playlists', 'in_deezer_playlists', 'in_shazam_charts']

# Clean and convert string numbers with commas to actual integers
for col in features:
    spotify_df[col] = spotify_df[col].astype(str).str.replace(',', '').astype(float)

# Create a subset DataFrame
subset = spotify_df[features]

# Compute the correlation matrix
corr_matrix = subset.corr()

# Get correlations between Spotify playlist inclusion and the other platforms
playlist_corr = corr_matrix['in_spotify_playlists'].drop('in_spotify_playlists')

# Define color palette
blues = sns.color_palette("Blues", n_colors=len(playlist_corr))

# Plot the bar graph
plt.figure(figsize=(8, 6))
sns.barplot(x=playlist_corr.index, y=playlist_corr.values, palette=blues)
plt.axhline(0, color='gray', linestyle='--')
plt.ylabel('Pearson Correlation Coefficient')
plt.title('Correlation Between Spotify Playlist Count and Other Platforms')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#
features = ['in_spotify_playlists', 'in_apple_playlists', 'in_deezer_playlists', 'in_shazam_charts']

for col in features:
    spotify_df[col] = spotify_df[col].astype(str).str.replace(',', '').astype(float)


plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
sns.regplot(
    data=spotify_df,
    x='in_spotify_playlists',
    y='in_apple_playlists',
    scatter_kws={'alpha': 0.6},
    line_kws={'color': 'red'}
)
plt.title('Spotify vs Apple Playlist Inclusion')
plt.xlabel('Spotify Playlists')
plt.ylabel('Apple Playlists')


plt.subplot(1, 3, 2)
sns.regplot(
    data=spotify_df,
    x='in_spotify_playlists',
    y='in_deezer_playlists',
    scatter_kws={'alpha': 0.6},
    line_kws={'color': 'green'}
)
plt.title('Spotify vs Deezer Playlist Inclusion')
plt.xlabel('Spotify Playlists')
plt.ylabel('Deezer Playlists')

plt.subplot(1, 3, 3)
sns.regplot(
    data=spotify_df,
    x='in_spotify_playlists',
    y='in_shazam_charts',
    scatter_kws={'alpha': 0.6},
    line_kws={'color': 'blue'}
)
plt.title('Spotify vs Shazam Chart Appearances')
plt.xlabel('Spotify Playlists')
plt.ylabel('Shazam Charts')

plt.tight_layout()
plt.show()
```

```python

spotify_df['is_charted'] = (spotify_df['in_spotify_charts'] > 0).astype(int)
```

```python
charted = spotify_df[spotify_df['is_charted'] == 1]['energy_%']
not_charted = spotify_df[spotify_df['is_charted'] == 0]['energy_%']
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(charted, not_charted, alternative='greater')

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
alpha = 0.05
if p_value < alpha:
    print("✅ Charted songs have significantly higher energy.")
else:
    print("❌ No significant difference in energy.")
```

```python
sns.kdeplot(charted, label="Charted", fill=True)
sns.kdeplot(not_charted, label="Not Charted", fill=True)
plt.title("Energy Comparison: Charted vs Not Charted")
plt.xlabel("Energy (%)")
plt.legend()
plt.show()
```
